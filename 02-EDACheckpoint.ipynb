### Trend of Global Atmospheric CO₂ Over Time

In this section, we explore how global atmospheric CO₂ levels have changed over time. Using the monthly dataset, we visualize the trend of CO₂ concentrations to observe long-term increases and seasonal patterns.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(co2_monthly['decimal'], co2_monthly['average'])

plt.xlabel("Year")
plt.ylabel("CO₂ concentration (ppm)")
plt.title("Global Atmospheric CO₂ Trend (Monthly)")
plt.grid(True)

plt.show()
```

### Annual Average CO₂ Concentration

Next, we examine the annual average CO₂ concentrations. The annual dataset removes seasonal variation, making it easier to observe the long-term upward trend in atmospheric CO₂ levels.

```python
plt.figure(figsize=(10,5))
plt.plot(co2_annual['year'], co2_annual['mean'])

plt.xlabel("Year")
plt.ylabel("CO₂ concentration (ppm)")
plt.title("Annual Global Mean CO₂ Levels")
plt.grid(True)

plt.show()
```

### Seasonal Pattern of Atmospheric CO₂

This plot shows the seasonal pattern of atmospheric CO₂ levels across different months. By averaging the monthly values over all years, we can observe a repeating yearly cycle in CO₂ concentrations. These seasonal fluctuations are largely influenced by plant growth and carbon absorption during different seasons.

```python
plt.figure(figsize=(10,5))

monthly_avg = co2_monthly.groupby("month")["average"].mean()

plt.plot(monthly_avg.index, monthly_avg.values)

plt.xlabel("Month")
plt.ylabel("Average CO₂ (ppm)")
plt.title("Seasonal Pattern of Atmospheric CO₂")

plt.grid(True)
plt.show()
```

### Distribution of Monthly CO₂ Concentrations

This histogram shows the distribution of monthly atmospheric CO₂ concentrations. Most values fall within a similar range, reflecting the gradual increase of CO₂ levels over time. The distribution helps us understand the spread and variability of CO₂ concentrations in the dataset.

```python
plt.figure(figsize=(8,5))

plt.hist(co2_monthly["average"], bins=20)

plt.xlabel("CO₂ concentration (ppm)")
plt.ylabel("Frequency")
plt.title("Distribution of Monthly CO₂ Concentrations")

plt.show()
```
